```python
# data import
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#data modelling
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('data/heart_disease_data.csv')
data.head()
```

```python
print(len(data))
```

```python
X = data.drop(columns='target', axis=1)
y = data['target']
```

```python
X
```

```python
y
```

```python
xtrain,xtest, ytrain,ytest = train_test_split(X,y,test_size=0.2, stratify=y, random_state=2)
```

```python
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
'''
Model Eval train data

'''
x_train_pred = model.predict(xtrain)
train_data_acc = accuracy_score(x_train_pred, ytrain)
print('Accuracy on train data: {}'.format(train_data_acc))
```

```python
'''
Model Eval test data

'''

x_test_pred = model.predict(xtest)
test_data_acc = accuracy_score(x_test_pred, ytest)
print('Accuracy on test data: {}'.format(test_data_acc))
```

```python
'''
Building a predictive system
'''
input_data = (62,0,0,140,268,0,0,172,0,1.4,2,0,2)

input_data_arr = np.asarray(input_data)

#reshape data
input_data_reshaped = input_data_arr.reshape(1,-1)

pred = model.predict(input_data_reshaped)
print(pred)

if pred[0] == 0:
    print('Patient does not have heart disease')
else:
    print('Patient does have an heart disease')
```
