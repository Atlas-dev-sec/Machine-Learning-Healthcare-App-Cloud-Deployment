```python
# data import
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#data modelling
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('data/parkinsons_data.csv')
data.head(5)
```

```python
print(len(data))
```

```python
# separate features and target
X = data.drop(columns=['name','status'], axis=1)
y = data['status']

print(X)
print(y)
```

```python
xtrain,xtest, ytrain,ytest = train_test_split(X,y, test_size=0.2, random_state=2)

#standarize data
scaler = StandardScaler()

std_x_train = scaler.fit_transform(xtrain) 
std_x_test = scaler.fit_transform(xtest)

print(std_x_train)
print(std_x_test)
```

```python
'''
MODEL TRAINING

SVM MODEL

'''
model = svm.SVC(kernel='linear')

#training model
model.fit(std_x_train, ytrain)
```

```python
'''
Model EVAL Training data
Accuracy score
'''

x_train_pred = model.predict(std_x_train)
train_data_acc = accuracy_score(ytrain, x_train_pred)

print('Accuracy score of training data: ', train_data_acc)
```

```python
'''
Model EVAL Test data
Accuracy score
'''
x_test_pred = model.predict(std_x_test)
test_data_acc = accuracy_score(ytest, x_test_pred)

print("Accuracy score of test data: ", test_data_acc)
```
